```python
pd.set_option('display.max_columns', None)
data = pd.read_csv('/Users/willfitzhugh/Desktop/Coding/Supply-Chain-Analysis/Data/Interim/Ready_For_DeliveryTime_Features.csv',
                  parse_dates=['OrderDate'])
```

#### Dataset including some info on product category, order region and customer state 


```python
data = pd.read_csv('/Users/willfitzhugh/Desktop/Coding/Supply-Chain-Analysis/Data/Interim/Ready_For_DeliveryTime_Features.csv',
                  parse_dates=['OrderDate'])
data.head()
```

```python
# note some product category outliers are significantly more or less likely to be late
data.groupby('ProductCategory').mean()[['late']].reset_index().sort_values('late').tail()
```

```python
data['pc high late risk'] = (data.ProductCategory.isin(['Cameras','Lacrosse','Golf Bags & Carts'])
                            ).map({True:1,False:0})
data['pc low late risk'] = (data.ProductCategory.isin(['Computers','Baby',"Men's Golf Clubs"])
                            ).map({True:1,False:0})
```

```python
data.groupby('Order Region').mean()[['late']].reset_index().sort_values('late')
```

```python
data['or high late risk'] = (data['Order Region'].isin(['Central Africa'])
                            ).map({True:1,False:0})
data['or low late risk'] = (data['Order Region'].isin(['Canada'])
                            ).map({True:1,False:0})
```

```python
data.groupby('Customer State').mean()[['late']].reset_index().sort_values('late')
```

```python
data['cs high late risk'] = (data['Customer State'].isin(['NM'])
                            ).map({True:1,False:0})
data['cs low late risk'] = (data['Customer State'].isin(['AL','MT'])
                            ).map({True:1,False:0})
```

```python
data.to_csv('/Users/willfitzhugh/Desktop/Coding/Supply-Chain-Analysis/Data/Model/DeliveryData_2.0.csv', index=False)
```
